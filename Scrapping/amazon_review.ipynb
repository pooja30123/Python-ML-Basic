```python
for i in bigbox:
    product = "https://www.amazon.in" + i.div.span.a['href']
    pro_open = urlopen(product)
    product_req = pro_open.read()
    product_html = bs(product_req , 'html.parser')
    product_name = product_html.findAll("span",{"id":"productTitle"})[0].text
    print("PRODUCT NAME :",product_name)
    comment_box = product_html.find_all("div",{"data-hook":"review"})
    for i in comment_box:
        user = i.findAll("span",{"class","a-profile-name"})[0].text
        rating = i.findAll("i",{"data-hook":"review-star-rating"})[0].text
        review_title = i.findAll("span")[3].text
        comment = i.findAll("div",{"data-hook":"review-collapsed"})[0].text
        date = i.findAll("span",{"data-hook":"review-date"})[0].text
        print("USER NAME :",user)
        print("RATING :",rating)
        print("REVIEW TITLE :",review_title)
        print("COMMENT :",comment)
        print("DATE :",date ,"\n\n\n\n")
    
```

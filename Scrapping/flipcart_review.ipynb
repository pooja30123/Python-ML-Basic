```python
for i in range(len(bigbox)):
    product_link = "https://www.flipkart.com" + bigbox[i].div.div.div.a['href']
    product_page = requests.get(product_link)
    product_html = bs(product_page.text , 'html.parser')
    product_name = product_html.find_all("span",{"class":"B_NuCI"})[0].text
    product_price = product_html.find_all("div",{"class":"_30jeq3 _16Jk6d"})[0].text
    print(product_name)
    print(product_price,"\n\n\n\n")
    comment_box = product_html.find_all("div",{"class":"_16PBlm"})
    for i in comment_box:
        review_headline = i.find_all("p",{"class":"_2-N8zT"})[0].text
        comment = i.find_all("div",{"class":""})[0].text
        rating = i.find_all("div",{"class":"_3LWZlK _1BLPMq"})[0].text
        user_name = i.find_all("p",{"class":"_2sc7ZR _2V5EHH"})[0].text
        print(review_headline)
        print(comment)
        print(rating)
        print(user_name, "\n\n")    
```

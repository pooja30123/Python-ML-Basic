```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [1,2,3,4,5,6] , y = [1,4,9,16,25,36] , mode = 'markers'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [1,2,3,4,5,6] , y = [1,4,9,16,25,36] , mode = 'lines'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4,5,6] , y = [1,4,9,16,25,36] ))
fig.show()
```

```python
x = [1,2,6,5,5,6,5,6,4,2,2,2,3,2,6,8,6,8,5,9]
```

```python
fig = go.Figure(data = [go.Histogram(x = x)])
fig.show()
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
fig = go.Figure(data = [go.Histogram(x = tips.total_bill)])
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill , y = tips.tip , mode = 'markers'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill , y = tips.tip , mode = 'markers', marker_size = 5*tips['size']))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x= tips.total_bill , y = tips.tip , z = tips['size'], mode = 'markers'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x= tips.total_bill , y = tips.tip , z = tips['size'], mode = 'lines'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x= [1,2,3,4] , y = [5,2,3,2] , z = [6,5,4,2], mode = 'lines'))
fig.show()
```

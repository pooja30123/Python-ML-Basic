```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

# Plot

```python
x = np.linspace(0,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(8,6))    #set size of figure
plt.plot(x,y,'--r')          #plot with color eg y,b,r,g
plt.xlabel("This is my x data")
plt.ylabel("This is my y dara")
plt.title("This is Title")
plt.grid()                    #grid line
plt.show()
```

# Scatter

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
col = np.random.rand(50)
size = 2000*np.random.rand(50)
plt.scatter(x,y,c=col,s=size,alpha=.7,marker='*')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data")
```

# Bar

```python
x = ['a','b','c','d','e']
```

```python
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar Plot")
```

```python
y
```

```python
y = np.random.rand(5)
plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar Plot")
```

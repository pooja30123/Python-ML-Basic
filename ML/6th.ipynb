# Standardization

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill=list(df['total_bill'])
```

```python
total_bill
```

```python
sns.histplot(total_bill)
```

```python
import numpy as np
```

```python
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
normalized_data = []
for i in total_bill:
    z_score=(i-mean)/std
    normalized_data.append(z_score)

print(normalized_data)
```

```python
sns.histplot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit_transform(df[['total_bill']])
```

# Normalization Min Max

```python
df=sns.load_dataset('taxis')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max=MinMaxScaler()
```

```python
min_max.fit(df[['distance','fare','tip']])
```

```python
min_max.fit_transform(df[['distance','fare','tip']])
```

```python
min_max.transform(df[['distance','fare','tip']])
```

```python
min_max.transform([[1,2,3]])  #new data
```
